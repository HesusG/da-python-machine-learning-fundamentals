```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Se lee el data set
df = pd.read_csv('datasets/dataset1.csv') 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

df.plot.scatter(x='x',y='y')
```


Para este ejemplo la variable 'x' será nuestra variable independiente y la variable 'y' sera nuestra variable dependiente.

```python
x = np.asarray(df[['x']])
y = np.asarray(df[['y']])
```


Ahora instanciaremos el modelo lineal y entrenaremos de forma supervisada

```python
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,y_pred,'--r')
```

Score para el modelo de regresión lineal

```python
model.score(x,y)
```


Hagamos lo mismo pero con el modelo KNN

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x,y)
y_pred = knn.predict(x)
plt.scatter(x,y)
plt.plot(x,y_pred,'--r')
```

```python
knn.score(x,y)
```
